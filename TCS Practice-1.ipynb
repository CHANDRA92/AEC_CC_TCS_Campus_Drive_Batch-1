### **CATSDOGS**

```python
// https://www.codechef.com/PRACTICE/problems/CATSDOGS
// in C++
#include <iostream>
using namespace std;
int main() {
    long long int test_cases, cats, dogs, legs, min_legs, max_legs;
    cin >> test_cases;
    while (test_cases--) {
        cin >> cats >> dogs >> legs;
        max_legs = (cats + dogs) * 4;
        if (cats > (2 * dogs)) {
            min_legs = (cats - (dogs * 2) + dogs) * 4;
        } else {
            min_legs = dogs * 4;
        }
        if (legs % 4 == 0 && legs >= min_legs && legs <= max_legs) {
            cout << "yes" << endl;
        } else {
            cout << "no" << endl;
        }
    }
}
```

### **PIPSQUIK**

```python
# in C++
// https://www.codechef.com/practice/course/2-star-difficulty-problems/DIFF1500/problems/PIPSQUIK
#include <iostream>
#include <bits/stdc++.h>
using namespace std;
int main() {
    int test_cases;
    cin >> test_cases;
    while (test_cases--) {
        int n, h, y1, y2, l, d = 0;
        cin >> n >> h >> y1 >> y2 >> l;
        vector<pair<int, int>> v(n);
        for (int i = 0; i < n; i++) {
            int t, x;
            cin >> t >> x;
            v[i].first = t;
            v[i].second = x;
        }
        for (int i = 0; i < n; i++) {
            if (l == 0) break;
            if (v[i].first == 1) {
                cout << "Type-1: " << y1 << " , " << y2<< " , "  << l<< " , "  << d<< " , "  << v[i].second << endl; 
                if (h - y1 > v[i].second) l--;
            } else {
                cout << "Type-2: " << y1 << " , " << y2<< " , "  << l<< " , "  << d<< " , "  << v[i].second << endl; 
                if (y2 < v[i].second) l--;
            }
            d++;
        }
        if (l == 0) d--;
        cout << d << endl;
    }
}
```
